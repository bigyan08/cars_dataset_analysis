## Cars Dataset Analysis

```python
import pandas as pd
```

```python
df=pd.read_csv(r"cars.csv")
```

```python
### Cleaning the data by finding the null values and filling them with its mean
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
### for string values we replace it with "unknown"
```

```python
df.head()
```

```python
df = df.fillna({'Make': 'Unknown', 'Model': 'Unknown', 'Type': 'Unknown','Origin': 'Unknown', 'DriveTrain': 'Unknown'})
```

```python
df.iloc[:,5:15]
```

```python
### now we need to convert the string value of msrp and invoice to numeric and find the mean and fill the null values with the mean!
```

```python
msrp_float=df['MSRP'].replace('[\$,]','',regex=True).astype(float)
```

```python
msrp_mean=msrp_float.mean()
```

```python
df['MSRP'].fillna(msrp_mean, inplace=True)
```

```python
invoice_float=df['Invoice'].replace('[\$,]','',regex=True).astype(float)
```

```python
invoice_mean=invoice_float.mean()
```

```python
df['Invoice'].fillna(invoice_mean, inplace=True)
```

```python
### now taking rest of the columns with numeric value and filling the null values with the mean of the column
```

```python
df.isnull().sum()
```

```python
df.iloc[:, 7:15] = df.iloc[:, 7:15].fillna(df.iloc[:, 7:15].mean())
```

```python
df.isnull().sum()
```

```python
### here, the data cleaning is completed and all the null values has been filled with the mean of the each selector
```
